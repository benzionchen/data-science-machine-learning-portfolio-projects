** How many people have the job title of "Lawyer" ? **


```python
def lawyer_title(jobTitle):
    if "lawyer" in jobTitle.lower().split():
        return True
    else: 
        return False
```

```python
# writing a test case

lawyer_title("Chief")
```

```python
# writing a test case

lawyer_title("LAWYER")
```

```python
sum(ecom['Job'].apply(lambda x:lawyer_title(x)))
```

** How many people have American Express as their Credit Card Provider *and* made a purchase above $95 ?**

```python
ecom[ecom['CC Provider'] == "American Express"]

# this is all people who have American Express
```

```python
ecom[ecom['Purchase Price'] > 95.00].nunique()

# 505 people have made purchases above $95
```

```python
# this is conditional selection with multiple conditions

ecom[ecom['CC Provider'] == "American Express"][ecom['Purchase Price'] > 95.00].nunique()
```

** Hard: How many people have a credit card that expires in 2025? **

```python
# confirm whether the value inside the cell under the CC expiration date column is a string or an integer
# we can test this using the 0th row cell

ecom['CC Exp Date'].iloc[0]
```

```python
sum(ecom['CC Exp Date'].apply(lambda exp: exp[3:]=='25'))

# for every cell's value under the "CC Exp Date" column, we apply the lambda expression to only index from the 3rd index onward
# and see which yields true

# and from that, we sum all the numbers of "True"
```

```python
ecom[ecom['CC Exp Date'].apply(lambda exp: exp[3:]=='25')].count()
```
