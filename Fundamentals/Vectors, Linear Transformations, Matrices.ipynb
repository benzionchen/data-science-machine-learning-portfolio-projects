# Vectors

Vector is a list of numbers (that are ordered) to express data
- vectors start at the origin in linear algebra 
- first number tells you how far to walk along the x-axis (usually called `i-hat`, and second number (below the first number) tells you how far to walk along the y-axis (called `j-hat`)
    - `i-hat` and `j-hat` are called the basis of a coordinate system (it means that when you add vectors and they become scalars, these basis vectors are what they scale)
    - vectors are written vertically with square brackets
- what about 3D (now it's a triplet of number) 
    - every number gives you a vector in the vector space 



Squishing, changing the direction, stretching the vector (dividing the vector in size, chaning from negative -> positive or vice versa, and multiplying the vector) is called Scaling
- when you scale a vector it's called a "Scalar" (this is known as a linear combination of the 2 vectors)
- it's called linear because if you fix one of those scalars, and let the other change its value freely, the tip of the resulting vector will form a line 



If you let 2 scalars move freely, 3 possible things can happen:
1. for most pairs of vectors, you'll be able to reach every possible point in the plane
    - every 2d vector is within your grasp
    


2. in the unlucky case that your vectors directly overlap/line up with each other, the tip of the resulting vector is limited to just the single line passing through the origin



3. there is a third possibility too, which is the vectors can both be "0" and you would get stuck at the origin as a dot 

## Span

The set of all possible vectors that you can reach with a linear combination of a given pair of vectors is called the span of those two vectors  
- 2 vectors "v" and "w" is the set of all their linear combinations

a(v) + b(w) where "a" and "b" are real numbers

The span of most 2d vectors is all vectors of 2d space, but when they line up, their span is the all vectors whose tip sit on a certain line

In 3D space, 2 vectors in 3D space not pointing in the same direction... what does it mean to take their span? 
- their span is the collection of all possible linear combinations of the 2 vectors, meaning all possible vectors you get by scaling each of the two of them in some way and then adding them together 


- you can kind of imagine turning different knobs to change the different scalars that define the linear combination
- adding the scaled vectors and following the tip of the resulting vector
    - that tip will trace out a flat sheet cutting through the origin of the 3D Space 
    


- this flat sheet is the span of the 2 vectors (more precisely, the set of all possible vectors whose tips sit on the flat sheet)

What about the span of 3 of these vectors? 





Linear combination of 3 vectors is pretty much the same as 2
- you choose 3 different scalars, scale each of those vectors, and then add them all together
- the span of these vectors is the set of all possible linear combinations






- if your third vectors happens to be sitting on the span of the first two, span doesn't change (you're trapped on the same flat sheet)  
- but if you randomly choose a third vector, it's likely not going to be sitting in the span of the first two, and since it's pointing in a separate direction it unlocks access to every possible 3-dimensional vector 


In the case where the third vector was already sitting on the span of the first two, or the case where two vectors line up... the way we can describe that one of these vectors is redundant and not adding to our span is called **"Linearly Dependent"**
- one of the vecotrs can be removed and doesn't affect the span
    - a.k.a. one of the vectors is expressed as a linear combination of the others, since it's already in the span of the others 

###  Vectors vs. Points

Vectors by themselves look like arrows, but a collection of vectors, we like to think of them without their tails at the origin so looking at the tip of the vector pointing directly at you, it's a dot (point)


## Basis 

The basis of a vector space is a set of linearly independent vectors that span the full space. This makes sense because: 


- Linearly independent - set of vectors is linearly independent if no vector in the set can be written as a linear combination of other vectors (meaning there are no redundant vectors) 

- Span the full space - means that any vector in the space can be expressed as a linear combination of the vectors in the basis
    - if the set of vectors did not span the full space, there would be vectors in the space that you couldn't create from your basis vectors meaning that the set does NOT cover the entire vector space 
    
Basis is like the smallest complete set of building blocks - A recipe for constructing any vector in the space

Span is everything you can build with the blocks - Menu of all vecotrs you can create using that recipe
