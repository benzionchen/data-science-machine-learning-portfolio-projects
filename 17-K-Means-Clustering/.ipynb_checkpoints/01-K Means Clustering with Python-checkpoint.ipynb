# Introduction to K Means Clustering Theory

K Means Clustering is an unsupervised learning algorithm that tries to cluster data based on their similarity. Unsupervised learning means that there is no outcome to be predicted, and the algorithm just tries to find patterns in the data. 

In k means clustering, we have the specify the number of clusters we want the data to be grouped into. The algorithm randomly assigns each observation to a cluster, and finds the centroid of each cluster. Then, the algorithm iterates through two steps:
(1) Reassign data points to the cluster whose centroid is closest. (2) Calculate new centroid of each cluster. These two steps are repeated till the within cluster variation cannot be reduced any further. The within cluster variation is calculated as the sum of the euclidean distance between the data points and their respective cluster centroids.

Math behind this is Chapter 10 of Gareth James' Introduction to Statistical Learning.

Some examples of clustering problems: 
- similar documents
- customers based off of features
- market segmentation
- identify similar physical groups

The goal is to divide data into distinct groups and observations/data samples within the same group at similar. This takes place with unlabelled training data and K means clustering clusters data into color-coded groups



## How does the algorithm work?

The objective is to minimize the total within-cluster variation, which is measured by Sum of Squared Errors. 

Observation = 1 data point = 1 sample of the sample size = 1 "n" (small "n")

Step 1. Choose a number "K" - and there's a method of choosing "K"
- there's no easy answer for choosing best "K", one way is the elbow method 
    - first, compute the sum of squared error (SSE) for some values of "K" (for example: k=2,4,6,8,etc.)
        - SSE is defined as the sum of the squared distance between each member of the cluster and it's centroid 
        - if you plot "K" against the SSE, you will see that error decreases as "K" gets bigger because the number of clusters increasing causes distortion to be smaller as well 
            - SSE will also go naturally up if there are more datapoints 
            - when the number of clusters increases, it means the clusters themselves end up being smaller, which means distortion ends up being smaller
    - the point of the elbow method is to choose "K" at which SSE decreases abruptly 

   - "Number of Clusters" = "K"-value (number of clusters you choose for the algorithms)
   - "Within Groups Sum of Squares"-Error (WGSS or WSS) or "Within Cluster Sum of Squares"-Error (WCSS) represents the total variance within each cluster, measuring the compactness of the clusters and calculated by the sum of squared euclidean distance between each data point and the centroid of its respective cluster (the less compact means the bigger the error, a higher WSS value indicates farther distance from centroid which is interpreted as a higher error) 
        - it's how tightly grouped each cluster is, and the elbow method uses WSS to determine the optimal "K" clusters
        - note: Euclidean distance is the straight-line physical distance between every data point and the centroid, then square the distance so we can emphasize larger differences in euclidean space (making it more sensitive to outliers away from the rest of the cluster), AND avoids the need for square root operations in distance calculations (computational efficiency)
            - euclidean distance will be in the units that the data is in (if data in miles, euclidean space is miles)
    - What the graph above is indicating is AS WE HAVE MORE CLUSTERS ("K"-number of clusters) we can MINIMIZE "within-cluster variance" which is something we use to measure accuracy in the context of clustering
        - we cannot use traditional methods of measuring accuracy that would typically be used in supervised learning because clustering is UNSUPERVISED -- which means that there's no way to test predictions against "correct answers" (like actual y_test values)
        - if "K" is too low or high it might not accurately reflect the data, leading to poor clustering quality 
            - low = underfitting, high = overfitting (including outliers)




Step 2. Randomly assign each datapoint to a cluster (it'll be a mixture of random points)
- in iteration 1, step 2a, the cluster centroids are then computed (these are shown as large colored discs) 
    - the initial centroids are almost completely overlapping because the initial cluster assignments are chosen at random (random in the sense of irregardless of what cluster/group of data points)
- in iteration 1, step 2b, each observation/data point is assigned to the nearest centroid 
    - in this step, we are calculating the squared Euclidean distance between point and each centroid to assign each data point to the nearest centroid 
    - after, we compute SSE by summing up the squared dsistances for all data points within each cluster
- in iteration 2, step 2a, identify and assign new cluster centroids - we do this by taking average of all datapoints assigned to each of the centroid (previously), and moving the centroid closest to the average 
    - a bare bones example (without squaring distances):
        - say we have 1-dimensional line segment with value of 3 and 8, and the centroid is initially at point-value 3
        - we take the average of 3 and 8 (3+8)/2 = 5.5 
        - we move the centroid from its old position at 3 and move it to 5.5 as its new centroid 
        - basically think of this except for 2-dimensional data (where there is a height and length) in a grid AND we have to square it (in practice)
    - this shifts the centroids to more appropriate positions (being at the "center" of the cluster) -- it's not really center as in exact middle, but it's a weighted center meaning if more points are on the right, the centroid will be more on the right side, and if only 1 point is on the left, we're not going to position the centroid near the left since there's only 1 datapoint on the left
        - but it gets tricky because if the 1 datapoint happens to be an outlier, and is WAY FAR left, the algorithm will be sensitive to that, and will be thrown off since outliers squared are more impactful than compact clusters
        - what is the solution to this? we can remove the outlier, transform it, cap it, or separate the treatment of it (and run the K-Means clustering algorithm without the outlier and compare the results between with and without to see if the outlier really had a significant impact)

Step 3. Repeat these steps until there are no new clusters (meaning data points are not being reassigned to new cluster centroid) or if we see convergence (centroids no longer change significantly or max number of iterations are reached):
- for each cluster, compute the cluster centroid by taking the mean vector of points in the cluster
- assign each data point to the cluster for which the centroid is closest
- NOTE: you know you are done iterating when the clusters stabilize
    - what that means is when the centroids no longer change (and they remain the same after each iteration) - this often means centroids do not move more than a predetermined small threshold because the centroids represent the average mean position of all the points in the cluster 
    - if these averages are not changing, it implies the points no longer switch between clusters (i.e. from cluster A to cluster B to cluster C) from one iteration to the next
        - similarly, if every point stays in the same cluster after each iteration, that also just means it's impossible for the average of the cluster (centroid) to change positions
    - sometimes the dataset is so complex that we have to set a hard-stop on the number of iterations to prevent the algorithm from running indefinitely due to small oscillations or VERY slow convergence
    - another hard-stop could be based off of the little change happening between iterations for the "Within Cluster Square Sums"-Error (WCSS or WSS)
        - if we are seeing a marginal difference between the error amounts between iterations, it's a good time to stop

Final results
- after about 10 iterations, we have the final result

## Bias in Clustering Algorithms like K-Means

"Bias" in clustering algorithms that use Within-Cluster Sum of Squares (WCSS) is not the same as the same "bias" traditionally understood in statistical or machine learning terms. 

1. Euclidean bias - K-Means is meant to minimize the euclidean distance between points inside the clusters, which can make it biased towards spherical clusters of similar size, which means K-Means will perform well when there are natural spherical clusters in the data, and poorly when the data has irregular shapes or vastly different sizes
2. Initial centroid bias - the first placement of the centroids can greatly affect the final clustering outcome, so "k-means++" aims to reduce this by spreading out initial centroids, but there can still be randomness involved that lead to different results (especially with complex datasets)
3. Outlier sensitivity - WCSS is sensitive to outliers because squaring the distance can greatly impact the outliers on the total sum, and a few points far away from the centroid can disproportionately influence the cluster assignments and centroid's locations 
    - in supervised learning (like regressive models), an outlier can spike the slope (mx) or intercept (b) in the linear regression but K-Means is sensitive to outliers in a different way: the impact of outliers distorts the literal shape of the clusters rather than affecting the output 
    - ex: Elon walks into a bar and the average income in the bar is well in the billions 
4. Equal Size bias - due to nature of WCSS, K-means can be biased towards creating clusters of roughly equal numbers of observations even if the natural distribution of clusters is unbalanced, and this can cause misleading clusters when actual groups are significantly different sizes 

### In Supervised Learning (e.g., regression, classification):

Bias refers to the error due to overly simplistic assumptions in the learning algorithm. High bias can cause the model to miss relevant relations between features and target outputs (underfitting).
Variance refers to the error due to too much complexity in the learning algorithm. High variance can cause the model to model the random noise in the training data, rather than the intended outputs (overfitting).
In this context, there is a trade-off: typically, increasing the complexity of the model will decrease bias but increase variance, and vice versa. The goal is often to find a good balance between bias and variance, minimizing total error.

### In Unsupervised Learning (e.g., clustering like K-means):

Bias in this context doesn't correspond directly to the same concept in supervised learning. However, you could consider an algorithm 'biased' towards certain kinds of solutions based on its assumptions (e.g., K-means assumes clusters are spherical and evenly sized).
Variance could refer to the sensitivity of the algorithm to different initializations or subsets of the data. For instance, different runs of K-means can lead to different outcomes due to random initial cluster assignments.

## Considerations for balancing "K"-value without Elbow Method

K-means clustering is unsupervised learning, which means it aims to partition "n"-observations ("n" is known as samples or datapoints) in "K"-number of clusters without labelled data. Each observation belongs to a cluster of observations with the nearest mean, serving as a prototype/initial cluster. 

How does the algorithm determine partitioning? 
- we use euclidean distance in the euclidean space to determine distance measurements
- after all the points are assigned to a cluster, it represented a partitioning of the dataset where each point is closer to its own cluster's centroid than any others (creating a Voronoi diagram of the space) 
- we can also use the `Silhouette Score` which measures how similar an object is to its own cluster compared to other clusters 
    - this score assesses the distance between resulting clusters (higher score = better defined clusters)

Why not just throw efficiency out the window and just go for the highest possible "K"-value (1000 data points, let there be 1000 clusters!)? 
- the goal of clustering is to find underlying patterns in data, if every data point was its own cluster, you wouldn't find any patterns
- second, there would be severe overfitting, memorizing the data rather than learning from it (this is something we see similarly in prediction models where it captures noise/outliers along with the signal) -- this means your clustering will not generalize well into new data
- also, computational cost drastically increases with higher "K"-value (as usual) 
- Silhouette score would be useless because there is no intra-cluster or inter-cluster disances when each cluster is one point
- it's not practical in real life because it might not make sense that there are 1000 different car markets, and only 1 car seller in each market (for example)
    - it's not going have information worth extrapolating, and you cannot make decisions off of this which is the typical point of a clustering model

## Model Evaluation (Silhouette score)

A score used to assess the quality of clusters created by a clustering algorithm like K-means. 
- provides a way to determine how well each object lies within its cluster, which is a reflection of both the cohesion of the cluster itself and the separation from other clusters

# K Means using Elbow Method to find optimal "K"


```python
# Sample data is data[0]

WCSS_error = []

for i in range(1, 11):  # Test up to 10 clusters
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(data[0])
    WCSS_error.append(kmeans.inertia_)  # inertia_ is the WCSS

# Plotting the results onto a line graph to observe the 'Elbow'
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), WCSS_error, marker='o')
plt.title('The Elbow Method')
plt.xlabel('"K"-Number of clusters')
plt.ylabel('WCSS Error Rate')
plt.show()
```

`wcss = []:`
- This line initializes an empty list called wcss. WCSS stands for Within-Cluster Sum of Squares. This list will be used to store the WCSS values for different numbers of clusters.

`for i in range(1, 11):`
- This line starts a loop that will execute its body for i values from 1 to 10. Each i represents a different number of clusters (K) to be tested with K-means clustering.

`kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0):`
- This line creates an instance of the KMeans class from Scikit-learn's cluster module.

`n_clusters=i:` 
- Sets the number of clusters to the current value of `i` in the loop.

`init='k-means++'` 
- Chooses the initialization technique for locating the initial cluster centers. 'k-means++' is an improved method for centroid initialization that tends to result in better clusterings.

`max_iter=300` 
- Sets the maximum number of iterations of the K-means algorithm for a single run. This is the max number of times the algorithm will move all centroids before it stops.

`n_init=10` 
- Number of time the k-means algorithm will be run with different centroid seeds. The final results will be the best output of `n_init` consecutive runs in terms of inertia.

`random_state=0` 
- Ensures reproducibility by setting a seed to the random number generator.

`kmeans.fit(X):`
- This line runs the K-means clustering algorithm on your dataset (in our case above is `data[0]`). The `.fit()` method computes K-means clustering algorithm (fitting the algorithm onto the data)

`wcss.append(kmeans.inertia_)`
- After fitting K-means to the dataset, .inertia_ attribute of the kmeans object gives the WCSS for the model. This line appends that value to the wcss list. The WCSS is a measure of how internally coherent the clusters are.

`plt.figure(figsize=(10, 6))`
- This line creates a new figure for plotting with a specified size (10 inches by 6 inches).

`plt.plot(range(1, 11), wcss, marker='o')`
- This line plots the WCSS values stored in wcss against the number of clusters (1 through 10).

`range(1, 11) creates the x-axis values (1 to 10)`
- `wcss` is the y-axis values.
- `marker='o'` specifies that each point on the plot should be marked with a circle.

`plt.title('The Elbow Method')`
- Sets the title of the plot to 'The Elbow Method'.

`plt.xlabel('Number of clusters (K)')`
- Sets the label for the x-axis to 'Number of clusters (K)'.

`plt.ylabel('WCSS')`
- Sets the label for the y-axis to 'WCSS'.

`plt.show()`
- This line displays the plot. When you run this code in a Python environment, a window should open displaying the plot with the WCSS for different numbers of clusters.
