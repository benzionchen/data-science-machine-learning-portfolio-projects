## K Means Cluster Creation

Now it is time to create the Cluster labels!

** Import KMeans from SciKit Learn.**

```python
from sklearn.cluster import KMeans
```

** Create an instance of a K Means model with 2 clusters.**

```python
KMeans_cluster = KMeans(n_clusters=2)
```

**Fit the model to all the data except for the Private label.**

```python
KMeans_cluster.fit(college_df.drop('Private', axis=1))
```

** What are the cluster center vectors? (aka Centroids)**

```python
KMeans_cluster.cluster_centers_

# this is a 2 by 17 feature (dimensions) array which means there are 2 centroids with 17 dimensions attached to each one
# so the first 17 numbers is NOT "X" and the second 17 numbers is NOT "Y" -- they are not coordinates 
```

## Evaluation

There is no perfect way to evaluate clustering if you don't have the labels, however since this is just an exercise, we do have the labels, so we take advantage of this to evaluate our clusters, keep in mind, you usually won't have this luxury in the real world.

** Create a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.**

```python
# if you just want to add a column, just write the "dataframe_name['Column Name'] = "

college_df['Cluster'] = college_df['Private'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
# lecture notes

def converter(private):
    if private == "Yes":
        return 1
    else: 
        return 0

college_df['Cluster'] = college_df['Private'].apply(converter)
```

```python
college_df.head(1)
```

** Create a confusion matrix and classification report to see how well the Kmeans clustering worked without being given any labels.**

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(college_df['Cluster'],KMeans_cluster.labels_))
print('\n')
print(classification_report(college_df['Cluster'],KMeans_cluster.labels_))

# we pass in the "Clusters" column because we converted "Private" into "1" or "0" values 
# "Cluster" column is our "y_test" because "Cluster" was converted from the true values of "Private"-column
# which means that every column EXCEPT for "Private" was our "X_test"
```

Not so bad considering the algorithm is purely using the features to cluster the universities into 2 distinct groups! Hopefully you can begin to see how K Means is useful for clustering un-labeled data!

- in most cases, you can't see classification and confusion matrix since you don't have the labels (if you had the labels already, we wouldn't need to use unsupervised method

`KMeans.labels_` is an attribute that stores the labels of the data points after the algorithm has been fitted to the data
- these labels represent the index of the cluster each data point is assigned to by the K-means algorithm

```python
# perform elbow method for a better "K"-value and see if we can increase the "accuracy"
from sklearn.cluster import KMeans

WCSS_error = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(college_df.drop('Private', axis=1))  # Assuming college_df is your data frame and 'Private' is a column to drop
    WCSS_error.append(kmeans.inertia_)  # Note the lower-case 'kmeans', which refers to the instance

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), WCSS_error, marker='o')
plt.title('The Elbow Method')
plt.xlabel('"K"-Number of clusters')
plt.ylabel('WCSS Error Rate')
plt.show()
```

```python
KMeans_cluster = KMeans(n_clusters=5)
KMeans_cluster.fit(college_df.drop('Private', axis=1))

print(confusion_matrix(college_df['Cluster'],KMeans_cluster.labels_))
print('\n')
print(classification_report(college_df['Cluster'],KMeans_cluster.labels_))
```

Mismatch Between Clustering and Classification: You are trying to use clustering (an unsupervised technique) followed by metrics used typically for classification (a supervised technique). confusion_matrix and classification_report are meant for evaluating classification algorithms where the outcomes are known and binary or multi-class, not for evaluating clustering where the groupings might not align with actual categories in the data.

Warnings Explanation: The warnings are because in the labels predicted by KMeans (which are just cluster numbers, not actual classes), there might be clusters for which there are no corresponding true labels in your college_df['Cluster']. This leads to situations where recall, F-score, and precision cannot be calculated because they involve dividing by zero (e.g., if there are no true positives for a particular class). This is what "Recall and F-score are ill-defined and being set to 0.0 in labels with no true samples" means.

Using KMeans for Binary Classification Evaluation: The typical use of KMeans is not for direct binary classification tasks. If college_df['Cluster'] contains binary labels (e.g., representing Private or not Private), and you are trying to evaluate KMeans with 5 clusters against this binary classification, there's a fundamental mismatch. KMeans doesn't know anything about your binary labels; it's just trying to group the data into 5 clusters based on distances.

Solution: If your objective is binary classification (predicting Private vs. Not Private), then:

Use a binary classification algorithm (like logistic regression, decision trees, SVM, etc.) instead of KMeans.
If you want to use KMeans results for some reason (which is unconventional for binary classification), you would need to map the 5 clusters back to two groups somehow, but this requires domain knowledge or additional steps and usually does not make sense unless you have a very specific hypothesis about the clusters.
Number of Clusters: If you were experimenting with KMeans as part of exploratory data analysis, that's fine, but trying to directly compare its clusters to binary labels with classification metrics is not appropriate. The number of clusters in KMeans doesn't need to match the number of classes in your target variable for binary classification; these are separate issues. But using metrics like confusion_matrix requires that the number of groups being compared should align, or else the metrics won't make sense.

Addressing Warnings: If for some reason you need to continue with this approach (which I don't recommend for binary classification evaluation), and you just want to address warnings, you could handle zero-division cases in classification_report by setting zero_division=1 (or 0, depending on your preference) within the function call. But again, this doesn't resolve the conceptual mismatch between clustering evaluation and classification evaluation.

In summary, if you're working on a binary classification problem, it's advisable to use a classification algorithm rather than KMeans, and evaluate it with confusion_matrix and classification_report. If using KMeans, it should be more for data exploration rather than direct binary classification evaluation.
