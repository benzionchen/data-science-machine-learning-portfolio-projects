# Matplotlib Overview Lecture

## Introduction

Matplotlib is the "grandfather" library of data visualization with Python. It was created by John Hunter. He created it to try to replicate MatLab's (another programming language) plotting capabilities in Python. So if you happen to be familiar with matlab, matplotlib will feel natural to you.

It is an excellent 2D and 3D graphics library for generating scientific figures. 

Some of the major Pros of Matplotlib are:

* Generally easy to get started for simple plots
* Support for custom labels and texts
* Great control of every element in a figure
* High-quality output in many formats
* Very customizable in general

Matplotlib allows you to create reproducible figures programmatically. Let's learn how to use it! Before continuing this lecture, I encourage you just to explore the official Matplotlib web page: http://matplotlib.org/

# Matplotlib Object Oriented Method
Now that we've seen the basics, let's break it all down with a more formal introduction of Matplotlib's Object Oriented API.

This means we will instantiate figure objects and then call methods or attributes from that object.

## Introduction to the Object Oriented Method

The main idea in using the more formal Object Oriented method is to create figure objects and then just call methods or attributes off of that object. This approach is nicer when dealing with a canvas that has multiple plots on it. 

For more information on how big figure's axes sizes are, check out the documentation: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html

To begin we create a figure instance. Then we can add axes to that figure:

```python
# Create Figure objects (also known as like an empty/blank canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # takes in 4 arguments: left, bottom, width, height (range 0 to 1)
```

```python
# the arguments of plotting MUST be in the same cell as the blank canvas or it will not show a plotted visual (seen below):

axes.plot(x, y, 'b')
axes.set_xlabel('Set X Label')
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')

# see how it just says "Text(0.5, 1.0, 'Set Title')"?
```

```python
# this is the OOP method:

fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y, 'b')
axes.set_xlabel('Set X Label')
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```

#### Code is a little more complicated, but the advantage is that we now have full control of where the plot axes are placed, and we can easily add more than one axis to the figure:

```python
# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # insert set of axes

# 0.2 is 20% from the left (inward) 
# 0.5 is 50% from the bottom left corner from the (0,0)
# 0.4 is referring to 40% of the canvas size in width 
# 0.3 is 30% of the canvas size in height
```

```python
# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # insert set of axes

# 0.2 is 20% from the left (inward) 
# 0.5 is 50% from the bottom left corner from the (0,0)
# 0.4 is referring to 40% of the canvas size in width 
# 0.3 is 30% of the canvas size in height

# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_LABEL_AXES 1')
axes1.set_ylabel('Y_LABEL_AXES 1')
axes1.set_title('Larger Plot')

# Insert Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Smaller Plot');
```

# .subplots()

Subplots in OOP method

The plt.subplots() object will act as a more automatic axis manager on top of plt.figure

plt.subplots() basically does the "axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) and axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])" for you automatically that you would typically need to do when you write plt.figure() 


Examples of use cases:

```python
# this is similar to plt.figure() except use tuple unpacking to grab fig and axes
# rather than doing fig = plt.figure, we are tuple unpacking and doing "fig, axes" - the reason being is because "axes" is just
# an array object of matplotlib axes objects (you can iterate through it)
fig, axes = plt.subplots()

# use the axes object to add stuff to plot
axes.plot(x, y, 'b')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');

# where subplots differs is that we can specify number of rows and columns
```

#### you can specify the number of rows and columns when creating the subplots() object:

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

# empty canvas of 1 by 2 subplots
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3)

plt.tight_layout() # this will fix the formatting so that none of them overlap
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

plt.tight_layout()
```

```python
# axes is an array of axes to plot on

axes

# "axes" is just an array object of matplotlib axes objects (you can iterate through it)
```

#### We can iterate through this array:

- also note that because you can interate through it, you can also index it


```python
for current_ax in axes:
    current_ax.plot(x, y, 'b')
    current_ax.set_xlabel('x')
    current_ax.set_ylabel('y')
    current_ax.set_title('title')

# diplay the figure object:    
fig
```

```python
# example of indexing through axes array:

fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First Title')

axes[1].plot(y,x)
axes[1].set_title('Second Title')
```

A common issue with matplolib is overlapping subplots or figures. We can use **fig.tight_layout()** or **plt.tight_layout()** method, which automatically adjusts the positions of the axes on the figure canvas so that there is no overlapping content:

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

fig    
plt.tight_layout()
```

# Figure size, aspect ratio and DPI

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the `figsize` and `dpi` keyword arguments. 
* `figsize` is a tuple of the width and height of the figure in inches (on the jupyter notebook, and zooming in and out of the webpage does affect the sizing of the figsize visually on the local machine)
* `dpi` is the dots-per-inch (pixel per inch). 

For example:

```python
fig = plt.figure(figsize=(8,4), dpi=100)
```

```python
# another example:

fig = plt.figure(figsize=(3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

#### The same arguments can also be passed to layout managers, such as the `subplots` function:

```python
fig, axes = plt.subplots(figsize=(13,3))

axes.plot(x, y, 'b')
axes.set_xlabel('x-axis')
axes.set_ylabel('y-axis')
axes.set_title('TITLE');
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

Now that we have covered the basics of how to create a figure canvas and add axes instances to the canvas, let's look at how decorate a figure with titles, axis labels, and legends.

**Figure titles**

A title can be added to each axis instance in a figure. To set the title, use the `set_title` method in the axes instance:

```python
ax.set_title("title");
```

**Axis labels**

Similarly, with the methods `set_xlabel` and `set_ylabel`, we can set the labels of the X and Y axes:

```python
ax.set_xlabel("x_label_string")
ax.set_ylabel("y_label_string");
```

### Legends

You can use the **label="label text"** keyword argument when plots or other objects are added to the figure, and then using the **legend** method without arguments to add the legend to the figure: 

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x squared")
ax.plot(x, x**3, label="x cubed")

ax.legend() 
# in order to do ax.legend(), you need to make sure that the label text is found in the argument
```

Notice how the legend overlaps some of the actual plot!

The **legend** function takes an optional keyword argument **"loc= "** that can be used to specify where in the figure the legend is to be drawn. The allowed values of **"loc= "** are numerical codes for the various places the legend can be drawn. See the [documentation page](http://matplotlib.org/users/legend_guide.html#legend-location) for details. Some of the most common **loc** values are:

```python
# Lots of options....

ax.legend(loc=1) # upper right corner
ax.legend(loc=2) # upper left corner
ax.legend(loc=3) # lower left corner
ax.legend(loc=4) # lower right corner

# .. many more options are available

# Most common to choose
ax.legend(loc=0) # let matplotlib decide the optimal location
fig
```

```python
# just playing around with the legend location...


fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x squared")
ax.plot(x, x**3, label="x cubed")

ax.legend(loc=10) 
# look at how the legend is at the center? well that's because "loc=10" is the "center" in the website documentation (above)
```

# Setting colors, linewidths, linetypes and how the plot looks

Matplotlib gives you *a lot* of options for customizing colors, linewidths, and linetypes. 

There is the basic MATLAB like syntax (which I would suggest you avoid using for more clairty sake:

### Colors with MatLab-like syntax

With matplotlib, we can define the colors of lines and other graphical elements in a number of ways. First of all, we can use the MATLAB-like syntax where `'b'` means blue, `'g'` means green, etc. The MATLAB API for selecting line styles are also supported: where, for example, 'b.-' means a blue line with dots:

```python
# MATLAB style line color and style 
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # blue line with dots
ax.plot(x, x**3, 'g--') # green dashed line
```

## Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color = '#FF8C00')

# color parameter can take in multiple arguments, can take in stuff like "green", "blue", etc. and accepts as a string
# it can also take in RGB hex codes like #FF8C00
```

#### We can also define colors by their names or RGB hex codes and optionally provide an alpha value using the `color` and `alpha` keyword arguments. "alpha= " indicates opacity.

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5) # half-transparant
ax.plot(x, x+2, color="#8B008B")        # RGB hex code
ax.plot(x, x+3, color="orange")        # orange color
```

### Line and marker styles

To change the line width, we can use the `linewidth` or `lw` keyword argument. The line style can be selected using the `linestyle` or `ls` keyword arguments:

- "linewidth= 1" is default and if you want double the width, it would be "lw= 2"

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options: ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="green", lw=3, linestyle='-') # this is dashed line
ax.plot(x, x+6, color="green", lw=3, ls='-.') # this is dash point dash point line 
ax.plot(x, x+7, color="green", lw=3, ls=':') # smaller dashes
# "steps" is just a staircase

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', etc. (these arguments are passed as a string)
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+') # these are plus sign markers that don't show up too well
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o') # there are o's
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s') # this is square  
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=4) # markersize is an argument to make it bigger
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='*', markersize=8)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='+', markersize=16, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=12, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="red"); # "s" is for square I think? 
# you can customize the marker face color, the edge of the marker, the width of the edge, the edge color, etc. 

# marker codes are in the documentation
```

## Control over axis appearance

In this section we will look at controlling axis sizing properties in a matplotlib figure.

## Plot range

We can configure the ranges of the axes using the `set_ylim` and `set_xlim` methods in the axis object, or `axis('tight')` for automatically getting "tightly fitted" axes ranges:

```python
# let's say we only wanted the line plot between 0 and 1 on the x-axis, we can call our axix and use ".set_xlim()"

import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2

fig, ax = plt.subplots(figsize=(5,5))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color= "purple", lw= 2, ls= "--")

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range");
```
