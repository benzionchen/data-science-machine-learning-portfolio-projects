** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
fig3 = plt.figure()

fig3_axis1 = fig3.add_axes([0,0,1,1])
fig3_axis2 = fig3.add_axes([0.2,0.5,0.4,0.4])

fig3_axis1.plot(x,z,'b')
fig3_axis2.plot(x,y,'b')

fig3_axis2.set_xlabel('X')
fig3_axis2.set_ylabel('Y')
fig3_axis2.set_title('zoom')
fig3_axis2.set_ylim([30,50])
fig3_axis2.set_xlim([20.0,22.0])

fig3_axis1.set_xlabel('X')
fig3_axis1.set_ylabel('Z') # even though it's "x" and "z" axes, we use "set_ylabel" since "z" is on the y-axis of the graph
```

```python
# solution below:

fig3 = plt.figure()

axes_set1 = fig3.add_axes([0,0,1,1])
axes_set2 = fig3.add_axes([0.2,0.5,0.4,0.4])

# the large axes in the back
axes_set1.plot(x,z)
axes_set1.set_xlabel('X')
axes_set1.set_ylabel('Z')

# insert the X and Y axes plot inside of the large plot
axes_set2.plot(x,y)
axes_set2.set_title('zoom')
axes_set2.set_xlabel('X')
axes_set2.set_ylabel('Y')

# change the limits on the inserts
axes_set2.set_xlim([20,22])
axes_set2.set_ylim([30,50])
```

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
# in order to do subplots, need to do tuple unpacking

fig, axes = plt.subplots(nrows=1, ncols=2)

# since we're using subplots, we can call on the sequence or array of axes hence, "axes[0]" or "axes[1]"
axes[0].plot(x,y,"b--",lw=3)
axes[1].plot(x,z,"r",lw=3)
```

```python
# answer key: 

fig, axes = plt.subplots(1, 2)

axes[0].plot(x,y,color="blue",ls='--',lw=3)
axes[1].plot(x,z,color="red",lw=3)
```

** See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12,2))
# we want wider and short, so first number will be bigger than the second for "figsize=()"

axes[0].plot(x,y,"b",lw=3)
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")

axes[1].plot(x,z,"r--",lw=3)
axes[1].set_xlabel("x")
axes[1].set_ylabel("y")

plt.tight_layout()
```

```python
# answer key: 

fig, axes = plt.subplots(1, 2,figsize=(12,2))

axes[0].plot(x,y,color="blue",lw=3)
axes[1].plot(x,z,color="red",lw=3,ls='--')
```
